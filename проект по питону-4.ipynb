```python
from nltk.tokenize import word_tokenize

w1= word_tokenize(first_album.lower())
w2 = word_tokenize (second_album.lower())
w3 = word_tokenize (third_album.lower())
w4 = word_tokenize (fourth_album.lower())
all = [w1, w2, w3, w4]
print(all)
```

```python
def get_second_element(indexable_collection):
    return indexable_collection[1]
```

```python
words = []
spisok = []
symbols= '!``”“"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~—'
for album in all:
    for word in album:
        if word not in symbols and word not in english_stop:
            words.append(word)
    spisok.append(words)
    words = []
print(spisok[0])   #очистили от пунктуации и от стопслов (добавить в симвлоы)
        
        
```

```python

all_dict = []
for words in spisok:
    word2number = {}
    for i in words :
        if i not in word2number:
            word2number[i] = 1
        else:
            word2number[i] += 1 #посчитали сколько слов
    all_dict.append(sorted(word2number.items(),
       key=get_second_element,      #отсортировали по количеству слов
       reverse=True))   
print(all_dict)
```

```python
bagOfWordsA = spisok[0]
bagOfWordsB = spisok[1]
bagOfWordsC = spisok[2]
bagOfWordsD = spisok[3]
uniqueWords = set(bagOfWordsA + bagOfWordsB + bagOfWordsC + bagOfWordsD)

numOfWordsA = dict.fromkeys(uniqueWords, 0)
for word in bagOfWordsA:
    numOfWordsA[word] += 1
numOfWordsB = dict.fromkeys(uniqueWords, 0)
for word in bagOfWordsB:
    numOfWordsB[word] += 1
numOfWordsC = dict.fromkeys(uniqueWords, 0)
for word in bagOfWordsC:
    numOfWordsC[word] += 1
numOfWordsD = dict.fromkeys(uniqueWords, 0)
for word in bagOfWordsD:
    numOfWordsD[word] += 1

numOfWordsA
```

```python
def computeTF(wordDict, bagOfWords):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(bagOfWordsCount)
    return tfDict
     

tfA = computeTF(numOfWordsA, bagOfWordsA)
tfB = computeTF(numOfWordsB, bagOfWordsB)
tfC = computeTF(numOfWordsC, bagOfWordsC)
tfD = computeTF(numOfWordsD, bagOfWordsD)

tfA
```

```python
def computeIDF(documents):
    import math
    N = len(documents)
    
    idfDict = dict.fromkeys(documents[0].keys(), 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))
    return idfDict
     

idfs = computeIDF([numOfWordsA, numOfWordsB, numOfWordsC,numOfWordsD])

idfs
```

```python
def computeTFIDF(tfBagOfWords, idfs):
    tfidf = {}
    
    for word, val in tfBagOfWords.items():
        tfidf[word] = val * idfs[word]
    return tfidf
     
    
     
```

```python
tfidfA = computeTFIDF(tfA, idfs)
tfidfB = computeTFIDF(tfB, idfs)
tfidfC = computeTFIDF(tfC, idfs)
tfidfD = computeTFIDF(tfD, idfs)

s = sorted(tfidfA.items(),key=get_second_element,reverse=True )
t = sorted(tfidfB.items(),key=get_second_element,reverse=True )
f = sorted(tfidfC.items(),key=get_second_element,reverse=True )
r = sorted(tfidfD.items(),key=get_second_element,reverse=True )

df = pd.DataFrame(r)
df.head(11)
     
```
